##<font color = 'green'>input / print</font>

**input** nos permite ingresar valores para ser tomados por Python, mientras que **print** nos permite imprimir información en pantalla.

```python
nombre = input("Introduce tu nombre")
print('Hola' + nombre)
```

Podemos intercalar valores de variables al utilizar el comando print.

```python
'''
Vamos a solicitar ahora un dato más
'''

edad = input('Introduce tu edad') #solicitamos su edad
print(f'Hola, {nombre}!. Tienes {edad} años')
```

## ESTADISTICAS CON NUMPYArrays


```python
arry = np.array([1,3,4,5,6,7,23,62,91,36,71,12,37,21])
print(np.mean(arry)) # Calcula la media de los elementos del arry
print(np.median(arry)) # la mediana de los elementos ===
print(np.std(arry)) # la desviacion estandar de los elementos ===
print(np.var(arry)) # la varianza de los elementos ===
print(np.percentile(arry, 70)) # el percetil 70 o N Numero ===

```
